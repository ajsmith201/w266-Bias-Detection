# BLEU

```python
# import df here

df = pd.read_csv('500_epoch4.csv')
df = df.iloc[: , 1:]
df
```

```python
for index, row in df.iterrows():

    reference = [row['Actual Text'].split()]
    candidate = row['Generated Text'].split()
    df.loc[index, 'BLEU_new'] = sentence_bleu(reference, candidate)
```

```python
df
```

```python
print('Final BLEU_new: ', df['BLEU_new'].mean())
```
